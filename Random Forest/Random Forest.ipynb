<h1 style="color:black">Random Forest</h1>

**Importing Important Library**

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Importing the Dataset**

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

**Columns in Dataset**

```python
dir(digits)
```

**Plotting Digits using Matplotlib**

```python
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

**Creating a DataFrame from Data Column of Digits Dataset**

```python
df = pd.DataFrame(digits.data) #This is just the representation of Image in form of Data Matrix 
df.head()
```

**Adding Target Column to the DataFrame**

```python
df["target"] = digits.target
df
```

**Creating Dependent and Independent Variables**

```python
X = df.drop("target", axis="columns")
y = df.target
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
len(X_train),len(X_test)
```

**Creating Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

**Fitting the Model with Data**

```python
model.fit(X_train, y_train)
```

**Predicting the Values**

```python
y_predicted = model.predict(X_test)
```

**Score of Model**

```python
model.score(X_train, y_train)
```

**Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
