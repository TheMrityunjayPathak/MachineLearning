**Logistic Regression**

**Logistic regression is basically a supervised classification algorithm. In a classification problem, the target variable(or output), y, can take only discrete values for a given set of features(or inputs), X.**

**Just like Linear regression assumes that the data follows a linear function, Logistic regression models the data using the sigmoid function.**

**Logistic Regression is used when the dependent variable(target) is categorical.**

**Importing Libraries**

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

**Reading CSV File**

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

**Plotting a Scatter Plot for Visualization of Data**

*Here we can see that a elder person is more likely to buy the insurance as compared to a younger person*

```python
plt.scatter(df.age,df.bought_insurance,marker="+",color="red")
plt.xlabel("Age")
plt.ylabel("Bought Insurance")
```

**Creating a Train Test Split on Data**

```python
from sklearn.model_selection import train_test_split
```

**Creating Train Test Split with Test Size of 10% and Train Size of 90%**

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance,test_size=0.1)
```

**Creating a Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

**Fitting Data to the Model**

```python
model.fit(X_train,y_train)
```

**Predicting Values from Model**

```python
model.predict(X_test)
```

**Analyzing Score of Model**

```python
model.score(X_test,y_test)
```
